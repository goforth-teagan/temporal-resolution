```python
# iterate through data 
for i in range(len(data2)):
    plt.clf()
    z = data2[i]
    z = z.sort_values(by=['Region','Season',x2],ascending=[True, True, False])
    plt.plot(z[x2].values)
    
    z = z.sort_values(by=['Region','Season','Avg'],ascending=[True, True, False])
    plt.plot(z['Avg'].values)
    plt.xlabel('Hours')
    plt.ylabel(x2)
    plt.title(profiles[i])
    
    fig.savefig('../outputs/plots/'+x+'_'+profiles[i]+'_load_plot.png', bbox_inches='tight')
```

```python
# Sequence example plot 
seq_data = ['seq_4hr','seq_8hr','seq_12hr','seq_24hr']#,'seq_48hr','seq_120hr','bind_8760hr']

seq_data2 = []

for i in range(len(seq_data)):
    seq_data[i] = pd.read_csv('../outputs/'+x+'/'+x+'_8760_'+profiles[i]+'.csv')
    seq_data[i] = seq_data[i][seq_data[i]['R_Group']=='FRCC']
    seq_data2.append(seq_data[i])

plt.clf()
    
for i in range(len(seq_data2)):
    z = seq_data2[i]
    z = z.sort_values(by=['Region','Season',x2],ascending=[True, True, False])
    plt.plot(z[x2][0:48].values)

z = z.sort_values(by=['Region','Season','Avg'],ascending=[True, True, False])
plt.plot(z['Avg'][0:48].values)
plt.xlabel('Hours')
plt.ylabel(x2)

fig.savefig('../outputs/plots/seq_load_plot.png', bbox_inches='tight')
```
