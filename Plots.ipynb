# Difference Boxplots

```python
%matplotlib inline

## agg backend is used to create plot as a .png file
mpl.use('agg')

import matplotlib.pyplot as plt 
```

```python
data_plot1 = [IPM1['Diff'],IPM2['Diff']]
data_plot1_names = [IPM1_name,IPM2_name]

# Create a figure instance
fig = plt.figure()

# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1])

# Create the boxplot
bp = ax.boxplot(data_plot1)

## Custom x-axis labels
ax.set_xticklabels(data_plot1_names)

# Save the figure
fig.savefig('../outputs/plots/IPM_boxplot.png', bbox_inches='tight')
```
